```python
def s(n):
  lista_sn1 = []
  
  for i in range(1, n+1):
    lista_sn1.append(1/i)
  soma = sum(lista_sn1)
  return soma
```

```python
lista_n = []
lista_sn2 = []
for n in range(Nmin, Nmax + 1):
  lista_n.append(n)
  lista_sn2.append(s(n))
df = pd.DataFrame({"n": lista_n, "s(n)": lista_sn2})
df
```

```python
def a(n):
  lista_an1 = []
  
  for i in range(1, n+1):
    lista_an1.append(1/i)
  resultado = (sum(lista_an1)) - (np.log(n))
  return resultado
```

# a) e b)

```python
def fibo(n):
    if n==1:
        return 1
    elif n==2:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
```

```python
lista_n = []
lista_mn = []
for n in range(Nmin, Nmax + 1):
  lista_n.append(n)
  lista_mn.append(fibo(n))
df = pd.DataFrame({"n": lista_n, "m(n)": lista_mn})
df
```

```python
def m(n):
  if n == 1:
    return 1
  elif n == 2:
    return 1
  else:
    resultado = ((m(n-1) + m(n-2)) + 1)/(m(n-1) + m(n-2))
    return resultado
```
