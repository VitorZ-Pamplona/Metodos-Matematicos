##Questão 01


```python
kmin = int(input("Forneça o Kmin: "))
kmax = int(input("Forneça o Kmax: "))
m = int(input("Forneça o m: "))
tolerancia = float(input("Forneça a tolerância: "))

def b(n):
  lista_bn = []
  
  for i in range(1, n+1):
    lista_bn.append((-1)**i)
  soma = sum(lista_bn)
  return soma
```

##A)

```python
lista_k = []
lista_sk = []
for k in range(kmin, kmax + 1):
  lista_k.append(k)
  lista_sk.append(b(k))

plt.figure(figsize = (13, 6))
plt.plot(lista_k, lista_sk, "ro")
plt.show();
```
