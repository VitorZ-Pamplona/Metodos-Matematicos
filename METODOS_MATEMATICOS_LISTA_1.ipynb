```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from math import ceil
import math
%matplotlib inline
```

```python
print(
    """Escola a sequência:
    1 - Convergente
    2 - Sequência limitada e divergente
    3 - Sequência não-limitada
    4 - Sequência de limitação e convergência indefinida
    """)

sequencia = int(input("Sua escolha: "))

if sequencia == 1:
  def a(n): return 1 / (n + 1)
  
  funcao = " ln(n) / (e^n)"
  tolerancia = 0.001
  l = 0
  Nmin = 0
  Nmax = 10
  N = ceil((1 / tolerancia) - 1)
  lista_n = []
  lista_an = []
  for n in range(Nmin, Nmax + 1):
    lista_n.append(n)
    lista_an.append(a(n))
  df = pd.DataFrame({"n": lista_n, "a(n)": lista_an})
  print(df)
  
  # Plotagem
  plt.figure(figsize = (15, 8))
  plt.title(f"Para a(n) = {funcao}, limite = {l} || tolerancia = {tolerancia} || intervalo [{Nmin}, {Nmax}]", fontsize=22)
  plt.plot(lista_n, lista_an, "g*")
  plt.plot([Nmin, Nmax], [l - tolerancia, l - tolerancia], "b-")
  plt.plot([25, Nmax], [l, l], "y-")
  plt.plot([Nmin, Nmax], [l + tolerancia, l + tolerancia], "b-")
  plt.plot([0, 0], [-20,20], "r-")
  plt.show()

elif sequencia == 2:
  def a(n): return (-1)**n

  funcao = "(-1)**n"
  Mmin = -2
  Mmax = 2
  Nmin = 100
  Nmax = 120
  lista_n = []
  lista_an = []
  for n in range(Nmin, Nmax + 1):
    lista_n.append(n)
    lista_an.append(a(n))
  
  df = pd.DataFrame({"n": lista_n, "a(n)": lista_an})
  print(df)

  # Plotagem
  plt.figure(figsize = (15, 8))
  plt.title(f"Para a(n) = {funcao} || intervalo [{Nmin}, {Nmax}]", fontsize=22)
  plt.plot(lista_n, lista_an, "g-*")
  plt.plot([Nmin, Nmax], [Mmax, Mmax], "b-")
  plt.plot([Nmin, Nmax], [Mmin, Mmin], "b-")
  plt.plot([Nmin, Nmin], [Mmin,Mmax], "r-")
  plt.show()

elif sequencia == 3:
  def a(n): return math.log(2**n)

  funcao = "ln(2**n)"
  M = 2
  Nmin = 100
  Nmax = 120
  lista_n = []
  lista_an = []
  for n in range(Nmin, Nmax + 1):
    lista_n.append(n)
    lista_an.append(a(n))
  
  df = pd.DataFrame({"n": lista_n, "a(n)": lista_an})
  print(df)

  # Plotagem
  plt.figure(figsize = (15, 8))
  plt.title(f"Para a(n) = {funcao}, M = {M} || intervalo [{Nmin}, {Nmax}]", fontsize=22)
  plt.plot(lista_n, lista_an, "g-*")
  plt.plot([Nmin, Nmax], [M, M], "r-")
  plt.show()

else:
  def a(n): return (3*n)**2 + 2

  funcao = "3n + 2"

  Nmin = 100
  Nmax = 120
  lista_n = []
  lista_an = []
  for n in range(Nmin, Nmax + 1):
    lista_n.append(n)
    lista_an.append(a(n))
  
  df = pd.DataFrame({"n": lista_n, "a(n)": lista_an})
  print(df)

  # Plotagem
  plt.figure(figsize = (15, 8))
  plt.title(f"Para a(n) = {funcao} || intervalo [{Nmin}, {Nmax}]", fontsize=22)
  plt.plot(lista_n, lista_an, "g-o")
  plt.plot([Nmin,Nmax], [Nmin,Nmin], "r-")
  plt.show()



```
