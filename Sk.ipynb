```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
kmin = int(input("Forneça o Kmin: "))
kmax = int(input("Forneça o Kmax: "))
```

```python
def s(k):
  Lista_inf = []

  for i in range (1, k+1):
    Lista_inf.append(1/i**2)
  soma = sum(Lista_inf)
  return soma
```

```python
lista_k = []
lista_sk = []
for k in range(kmin, kmax+1):
  lista_k.append(kmin + k)
  lista_sk.append(s(k+1))
df = pd.DataFrame({"k": lista_k, "s(k)": lista_sk})
df
```
